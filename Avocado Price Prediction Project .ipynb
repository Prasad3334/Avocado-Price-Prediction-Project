```python
data.rename(columns={'4046':'T.Sold Avac PLU 4046','4225':'T.Sold Avac PLU 4025','4770':'T.Sold Avac PLU 4070'},inplace=True)
data.head()
```

```python
x=pd.get_dummies(data.drop(['Date','Averageprice','Months_Str','Unnamed: 0'],axis=1),columns=['Region','Year','Months'],drop_first=True)
x.head()
```

```python
y=data.Averageprice
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=13)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
col_std=['Total volume','T.Sold Avac PLU 4046','T.Sold Avac PLU 4025','T.Sold Avac PLU 4070','Total bags','Small bags','Large bags','Xlarge bags']
```

```python
st=StandardScaler()
```

```python
st.fit(x_train[col_std])
```

```python
x_train[col_std]=st.transform(x_train[col_std])
x_test[col_std]=st.transform(x_test[col_std])
```

```python
x_train
```

# Model Evaluation

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_lr=r2_score(y_test,y_pred)
r2_lr
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dc=DecisionTreeRegressor()
```

```python
dc.fit(x_train,y_train)
```

```python
y_pred_dc=dc.predict(x_test)
```

```python
r2_dc=r2_score(y_test,y_pred_dc)
r2_dc
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=10,random_state=13)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
r2_rf=r2_score(y_test,y_pred_rf)
r2_rf
```

```python
from sklearn.svm import SVR
```

```python
svm=SVR(kernel='linear',gamma='auto',C=10)
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred_svm=svm.predict(x_test)
```

```python
r2_svm=r2_score(y_test,y_pred_svm)
r2_svm
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
kn=KNeighborsRegressor(n_neighbors=10)
```

```python
kn.fit(x_train,y_train)
```

```python
y_pred_kn=kn.predict(x_test)
```

```python
r2_kn=r2_score(y_test,y_pred_kn)
r2_kn
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
evr=VotingRegressor(estimators=[('lr',lr),('dc',dc),('rf',rf),('svm',svm),('kn',kn)])
```

```python
model_evr=evr.fit(x_train,y_train)
```

```python
evr_pred=model_evr.predict(x_test)
```

```python
r2_score(y_test,evr_pred)
```

```python
list1=['Linear Regression','Decision Tree','Random Forest','Support Vector','K_Nearest_N']
list2=[r2_lr,r2_dc,r2_rf,r2_svm,r2_kn]
```

```python
Final=pd.DataFrame({'Method Used':list1,
                   'R2-Score':list2})
print(Final)
```
